*BRAND VS COUNT*

```python
ax = sns.countplot(x='brand', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

*Realme offers the most options, as seen by above bar graph. In terms of specs,they have a broad range of phones to select from.
Poco, on the other hand, has the fewest alternatives. This might be since Poco is a relatively new brand.*

```python
sns.set(rc={'figure.figsize':(9,4)})
ax = sns.countplot(x='processor', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

*SALES PRICE OF DIFFERENT BRAND FOR DIFFERENT COLOR*

```python
sales_state= df.groupby(['brand','base_color'],as_index=False)["sales_price"].sum().sort_values(by="sales_price", ascending=False)

sns.barplot(data=sales_state,x='brand', y='sales_price', hue='base_color')
```

*SALES COUNT OF DIFFERENT BRANDS FOR DIFFERENT COLOURS*

```python
sales_state= df.groupby(['brand','base_color'],as_index=False)["sales"].sum().sort_values(by="sales", ascending=False)

sns.barplot(data=sales_state,x='brand', y='sales', hue='base_color')
```

*SALES PRICE VS TOP 10 MODEL*

```python
sns.set(rc={'figure.figsize':(19,5)})
sales_state= df.groupby(['model'],as_index=False)["sales_price"].sum().sort_values(by="sales_price", ascending=False).head(10)

sns.barplot(data=sales_state,x='model', y='sales_price')
```
